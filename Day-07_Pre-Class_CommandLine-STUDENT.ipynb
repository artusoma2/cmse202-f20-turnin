# 2: Exploring using the Command Line interface

The Command Line Interface (CLI for short) is a powerful tool that can allow you to do many things from open files, to runnning programs, to even doing things that you normally would not be able to just using the Graphical User Interfce (GUI for short). The CLI is a realtivly straight forward thing to use once you get the hang of it. The bulk of the difficulty in starting to use the CLI is remembering what commands do what and how to move around without the assistance of a GUI. However, with a little practice, the commands become second nature and just another tool in the toolbelt.

The commands that you will use to naviagate around using CLI are the following:

## ls

* `ls` - Used to list the files/folders in your directory (fancy name for "folder") that you are in. This is the equivalent of looking in a folder in your File Explorer/Finder. In this case it will just be words instead of icons.
    * Example: Using `ls` and `ls -a` in the CLI next to the same folder being opened in File explorer.

## pwd

Meaning if you were to start at the beginning of your hard-drive, those are the folders you would have to travel through in order to get at where you are now. In other words the "path" to your current directory.

This is when you are typing a command or file/folder name and you can actually use the `tab` key on your keyboard to finish off the command or file/folder name. This should work almost everytime you are typing a file/folder name

# 3: Creating, Removing, Copying and Moving folders and files

Being able to manipulate files and folders in CLI is very useful tool, however, it is a task that can kind of be a little more diffiult to get used to coming from a GUI. Below we will cover the main maintence tasks you will need to be able to perform to manipulate files/folder in a CLI. 

## Creating directories and text files

*To create a folder (directory)* in the CLI you can use the command `mkdir [FOLDER_NAME]`.  In place of `[FOLDER_NAME]` is whatever you would like to name your folder. Note that when you create the folder, you will not automatically travel inside the folder. You will have to manually move inside the folder by changing directories into it. An example would be `mkdir example` and that would make a folder called `example`.

*To create a blank text file* in the CLI you can use the command `touch [FILENAME]`. In place of the `[FILENAME]` is whatever you would like to name your file. However, it is important to note that in addition to the filename you will also have to specify the file extension. An example of this would be `touch example.txt`. Technically you can create any filename with any extension, however keep in mind many programs rely on specific file extensions in order to use them. In the next section you will learn about text editors.

##  Copying folders and files

*To copy a file* in the CLI you can use the command `cp [FILENAME] [NEW_FILENAME]`. Where `[FILENAME]` is the file you want to copy and the `[NEW_FILENAME]` is the name you want the copy to be. Remember that the copy will also need to have the same extension as the original file. Example of this would be `cp copy.txt copy2.txt`

*To copy a folder* in the CLI is much like copying a file, however, you also need to use the `-r` flag. The `-r` flag means recursive" and means that you want to copy the folder and all of its contents. To copy a folder the command is `cp -r [FOLDER_NAME] [NEW_FOLDER_NAME]`. An example of this would be `cp -r folder1 folder2`.

*Note: That you do not have to be in the same directory as the item you are trying to copy. You can actually replace the `[FILENAME]`/`[NEW_FILENAME]` or  `[FOLDER_NAME]`/`[NEW_FOLDER_NAME]` with `[PATH_TO_FILE]`/`[PATH_TO_NEWFILE]` or `[PATH_TO_FOLDER]`/`[PATH_TO_NEW_FOLDER]`. Make sure when are doing this that name of the file is in the path as well. (This was not mentioned early, but just like you can use `.` in a path you can also use `..` as well). An example of this would be if you were in `~/CMSE202/example` and you wanted to copy `test.txt` to `~/CMSE202/example` you could use the command `cp test.txt ../scratch/test.txt`*

## Removing folders and files

*To remove a file* in the CLI you can use the commmand `rm [FILENAME]`. In place of `[FILENAME]` is whatever file you would like to delete. **Warning** once you delete a file there is no way to easy recover it. So be very careful in what files you delete. An example of this command would be `rm example.txt`

*To remove a folder* in CLI you can use the command `rm -r [FOLDER_NAME]`. Just like copy you need that `-r` tag. An example of this command would be `rm -r folder2/`

*Note: That you do not have to be in the same directory as the item you are trying to remove. You can actually replace the `[FILENAME]` or  `[FOLDER_NAME]` with `[PATH_TO_FILE]` or `[PATH_TO_FOLDER]`. But note that you will need to make sure you use the full path OR use the `.` or `..` shortcuts. An example of this would be if you were in `~/CMSE202/example` and you wanted to remove `scratch` you could use the command `rm -r ../scratch`*

## Moving folders and files

*To move a file/folder* is very similar to copying a file/folder, however, unlike with copying you *do not* need to use the `-r` flag in order to move a folder. The thing that might seem a little strange about moving a file/folder is you can actaully rename that file/folder as well as move it. The command for moving a file/folder is `mv [FILENAME/PATH_TO_FILE] [RENAMED_FILE/NEW_PATH_TO_FILE]` where the `[FILENAME/PATH_TO_FILE]` is the name or path of the file you want to move and the `[RENAMED_FILE/NEW_PATH_TO_FILE]` is the name you would like to rename the file to, or where you would like to move the file to (or both). An example of both moving a file from `example` to `scratch` and renaming is `mv test.txt ../scratch/test2.txt`.

## Vim



Vim is one of the harder CLI text editors, but also one of the most powerful, to use do to its bi-modal nature and lack of on-screen commands. Vim, however, is sometimes the only text editor avalible, or is the text editor used by default by CLI programs. It is important to know how to open, exit, and save in Vim. Also, it is important to know how to handle `.swp` files that are generated by Vim if it is exited improperly.

### Using Vim to edit text

As mentioned, Vim is what is known as bi-modal (meaning two-modes). There is a *command mode* where you can do actions like save and exit, exit without saving, delete many lines at once, temporarily exit Vim, and many other features. The other mode is *insert* mode which is the mode where you can actually edit the contents of the file and type whatever you would like. 

To start Vim in the terminal you will need to use one of two commands:
* `vim` - This will just open up vim in the equivalent of a "Untitled" document just like nano.
* `nano [FILENAME]` - This will open up the specified text file in vim. If the file does not exist, it will create that file as long as you save the file when closing out (just like nano).

When opening vim you will always start out in *command mode*. In *command mode*, you will be able to scroll up and down the text with the arrow keys (`←`,`↑`,`→`,`↓`), however, you will not be able to edit the text. To be able to edit the text you will need to change to *insert mode*. You can do this by pressing `i` on your keyboard. You should now see `-- INSERT--` on the bottom of the terminal.



In insert mode, you can edit the text however you see fit. Like nano you will need to use the arrow keys to navigate around (`←`,`↑`,`→`,`↓`). To return back to *command mode* you will need to pres `Esc` on your keyboard. You can tell that you are back to *command mode* as the `-- INSERT--` on the bottom of the terminal should be gone.

Now in order to save and exit you will need to type one of two commands
* `:wq` - To save and exit
* `:q!` - To exit and not save

**Using vim, open up the file you made with nano in `~/CMSE202/scratch`. View the file, but exit without saving. Then create a new file with Vim. Like above it can be any filename/extension you please. Save and Exit the file.**

### How to handle a `.swp` (swap file)

If forever reason, you exit Vim without running either of those two commands above to exit Vim, Vim will create what is known as a "swap file" (`.swp` file extension) of the file you were working on. If that is the case, when you try to open the file in Vim you will see something like below.



As you can see their was a swap file created for the file you were editing. You have a couple of different options you can do. In order to perform the action all you need to do is type the character, no need to even press `Enter`

* `o` - This will open the original file, before you entered the file, as a read only. 
* `e` - Will open the original file, before you entered the file, but you can make edits to it.
* `r` - This will change the original file to be the swap file, meaning if you made edits to file before it was closed improperly, this will replace the original file with that file.
* `d` - This will delete the swap file, allowing you to open the file as it stands
    * If you recovered the swap file already it will open the file as it was corrected earlier and you get rid of the swap file.
    * If you made no important edits to the file and you just want to delete the swap file use this option.
* `q` or `a` - This will exit this screen back to the CLI.

Note that in order to run the program without that warning you will need to delete the `.swp` file. This can be done either by choosing the `d` option above, or using `rm` to delete the file.
